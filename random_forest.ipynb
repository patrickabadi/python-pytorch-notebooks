```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np

np.random.seed(0)

```

```python
iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)

# print first five rows
df.head()
```

```python
# adding a new column for the species name (our target, what we're trying to predict)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

#view top five rows
df.head()
```

```python
# create training data
df['is_train'] = np.random.uniform(0, 1, len(df)) <= .75

# view the top 5 rows
df.head()
```

```python
# creating dataframes with test rows and training rows
train, test = df[df['is_train'] == True], df[df['is_train'] == False]

# show num observations each
print('Number of train data: ', len(train))
print('Number of test data: ', len(test))
```

```python
# create a list of the feature column names
features = df.columns[:4]

# view features
features
```

```python
# converting each species name into digits
y = pd.factorize(train['species'])[0]

y
```

```python
# creating a random forest classifier
clf = RandomForestClassifier(n_jobs=2, random_state=0)

# training the classifier
clf.fit(train[features], y)
```

```python
#applying the trained classifier to the test
clf.predict(test[features])
```

```python
test[features]
```

```python
# viewing the predicted probabilities of the first 10 observations
clf.predict_proba(test[features])[10:20]
```

```python
# mapping names for the plants for each predicted plant class
preds = iris.target_names[clf.predict(test[features])]

# view the predicted species for teh first five
preds[:25]
```

```python
# viewing the actual species for the first five observations
test['species'].head()
```

```python
# creating a confusion matrix
pd.crosstab(test.species, preds, rownames=['Actual species'], colnames=['Predicted species'])
```
